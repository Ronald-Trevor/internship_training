```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data_path = "C:/Users/CHALTECH/Downloads/museum_visitors.csv"
data = pd.read_csv(data_path,index_col="Date",parse_dates=True)
data.head()
```

```python
# Line chart showing the number of visitors to each museum over time
plt.figure(figsize=(12,6))
sns.lineplot(data=data)
plt.title("Monthly Visitors to Los Angeles City Museums")
plt.show()
```

```python
# Line plot showing the number of visitors to Avila Adobe over time
sns.lineplot(data=data['Avila Adobe'],label = 'Avila Adobe')
plt.xlabel = 'Date'
plt.show()
```

```python
data2_path = "C:/Users/CHALTECH/Downloads/candy.csv"
data2 = pd.read_csv(data2_path,index_col="id")
data2
```

```python
# Scatter plot showing the relationship between 'sugarpercent' and 'winpercent'
sns.scatterplot(x=data2['sugarpercent'],y=data2['winpercent'])
```

```python
# Scatter plot w/ regression line showing the relationship between 'sugarpercent' and 'winpercent'
sns.regplot(x=data2['sugarpercent'],y=data2['winpercent'])
```

```python
# Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate'
sns.scatterplot(x=data2['pricepercent'],y=data2['winpercent'],hue=data2['chocolate'])
```

```python
# Color-coded scatter plot w/ regression lines
sns.lmplot(x='pricepercent',y='winpercent',hue='chocolate',data=data2)
```

```python
# Scatter plot showing the relationship between 'chocolate' and 'winpercent'
sns.swarmplot(x=data2['chocolate'],y=data2['winpercent']) 
```
